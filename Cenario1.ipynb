```python
import pandas as pd

class Professor:
    def __init__(self, SERVIDOR, MATRICULA, COD_DISCIPLINA, TURNO):
        self.SERVIDOR = SERVIDOR
        self.MATRICULA = MATRICULA
        self.COD_DISCIPLINA = COD_DISCIPLINA
        self.TURNO = TURNO
        self.escolas_carga_horaria = {}  # Dicionário para associar escolas e carga horária

    def adicionar_escola_carga_horaria(self, escola, carga_horaria):
        self.escolas_carga_horaria[escola] = carga_horaria

# Carregar o DataFrame
base_professores = pd.read_excel('mock1.xlsx')
base_professores = base_professores[base_professores['COD_DISCIPLINA'] == 2002]

# Dicionário para armazenar os objetos Professor
professores_dict = {}

# Percorrer as linhas do DataFrame
for indice, linha in base_professores.iterrows():
    nome_objeto = f"{linha['SERVIDOR']}_{linha['MATRICULA']}_{linha['COD_DISCIPLINA']}"
        
    # Criar um objeto Professor para cada linha
    professor = Professor(
        linha['SERVIDOR'],
        linha['MATRICULA'],
        linha['COD_DISCIPLINA'],
        linha['TURNO']
    )

    # Adicionar o objeto ao dicionário usando o nome como chave
    professores_dict[nome_objeto] = professor

for nome_objeto, professor in professores_dict.items():
    print(f"Nome do objeto: {nome_objeto}")
```

```python

# Preencher as informações de escolas e carga horária
for nome_objeto, professor in professores_dict.items():
    for index, escolas in base_professores.iterrows():
        if professor.MATRICULA == escolas['MATRICULA']:
            if escolas['COD_ESCOLA'] not in professor.escolas_carga_horaria:
                professor.adicionar_escola_carga_horaria(escolas['COD_ESCOLA'], int(escolas['CH_TOTAL']))

# Imprimir as informações
print('-'*20)
for nome_objeto, professor in professores_dict.items():
    print(f"Nome do objeto: {nome_objeto}, Escolas e CH: {professor.escolas_carga_horaria}")
print('-'*20)

professores_dict_ordenado = {}

for nome_objeto, professor in professores_dict.items():
    sorted_escolas_carga_horaria = dict(sorted(professor.escolas_carga_horaria.items(), key=lambda item: item[1], reverse=True))
    professor.escolas_carga_horaria = sorted_escolas_carga_horaria
    professores_dict_ordenado[nome_objeto] = professor
    
print('-'*20)
for nome_objeto, professor in professores_dict.items():
    print(f"Nome do objeto: {nome_objeto}, Escolas e CH: {professor.escolas_carga_horaria}")
print('-'*20)
```

```python
import pandas as pd

colunas = [
    'Professor 1', 'Matricula 1',
    'Escola 1', 'Ch 1',
    'Escola 2', 'Ch 2',
    'Escola Destino', 'CH Destino',
    'Professor 2', 'Matricula 2',
    'Escola 3', 'Ch 3',
    'Escola 4', 'Ch 4',
    'Escola Destino Verificado', 'CH Destino Verificado',
]

# Inicializar o DataFrame
professores_trocados = pd.DataFrame(columns=colunas)

trocas_registradas = set()

for nome_objeto, professor in professores_dict.items():
    print(professor.escolas_carga_horaria)

for nome_objeto, professor in professores_dict.items():
    escolas_comp = dict(professor.escolas_carga_horaria.items())

    # Escola1 = list(professor.escolas_carga_horaria.keys())[0]
    # Escola2 = list(professor.escolas_carga_horaria.keys())[1]

    Escola1, Escola2 = escolas_comp

    ch_escola1 = professor.escolas_carga_horaria[Escola1]
    ch_escola2 = professor.escolas_carga_horaria[Escola2]
    ch_total = ch_escola1 + ch_escola2

    for objeto_verficado, prof_verificado in professores_dict.items():

        if nome_objeto != objeto_verficado:

            if set(escolas_comp) == set(prof_verificado.escolas_carga_horaria):

                Escola_verf1 = list(prof_verificado.escolas_carga_horaria.keys())[0]
                Escola_verf2 = list(prof_verificado.escolas_carga_horaria.keys())[1]

                ch_verf_escola1 = prof_verificado.escolas_carga_horaria[Escola_verf1]
                ch_verf_escola2 = prof_verificado.escolas_carga_horaria[Escola_verf2]

                nova_ch = ch_escola1 + ch_verf_escola2
                nova_ch_verif = ch_verf_escola1 + ch_escola2

                troca_atual = (
                    professor.SERVIDOR, prof_verificado.SERVIDOR,
                    Escola1, Escola2, Escola_verf1, Escola_verf2
                )

                if troca_atual not in trocas_registradas and Escola2 != Escola_verf2:
                    dados = {
                        'Professor 1': professor.SERVIDOR,
                        'Matricula 1': professor.MATRICULA,
                        'Escola 1': Escola1,
                        'Ch 1': ch_escola1,
                        'Escola 2': Escola2,
                        'Ch 2': ch_escola2,
                        'Escola Destino': Escola1,
                        'CH Destino': nova_ch,
                        'Professor 2': prof_verificado.SERVIDOR,
                        'Matricula 2': prof_verificado.MATRICULA,
                        'Escola 3': Escola_verf1,
                        'Ch 3': ch_verf_escola1,
                        'Escola 4': Escola_verf2,
                        'Ch 4': ch_verf_escola2,
                        'Escola Destino Verificado': Escola_verf1,
                        'CH Destino Verificado': nova_ch_verif,
                    }
                    
                    # Correção: use 'append', não '_append'
                    professores_trocados = professores_trocados._append(dados, ignore_index=True)

                    trocas_registradas.add(troca_atual)

# Exibir o DataFrame resultante
display(professores_trocados)
```
